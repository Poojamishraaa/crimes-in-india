```python
import pandas as pd
```

```python
# data collect from this website https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population
data1=pd.read_excel("population.xlsx")
```

```python
data1
```

```python

# Extract the state and population columns
state_population = data1[['STATE/UT', 'Population[18][19]']]

# Print the state and population information
print(state_population)
```

```python
#https://www.findeasy.in/indian-states-by-literacy-rate/
data2=pd.read_excel("lit2.xlsx")
data2
```

```python
# Extract the state and literacy rate columns
state_literacy_rate = data2[['STATE/UT', 'Literacy Rate']]

# Remove any NaN values
state_literacy_rate = state_literacy_rate.dropna()

# Print the state and literacy rate information
print(state_literacy_rate)
```
